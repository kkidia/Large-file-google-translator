```python
#lib
from pdf2docx import Converter
from deep_translator import GoogleTranslator
from docx import Document
from docx2pdf import convert
from PyPDF2 import PdfReader
```

# Fast translating but miss orignal fontsize

```python
pdf_path = r"C:\Users\Kidia.Gelaye\Downloads\Demo.pdf"
docx_path = r"C:\Users\Kidia.Gelaye\Downloads\Original_Formatted.docx"

cv = Converter(pdf_path)
cv.convert(docx_path, start=0, end=None)
cv.close()

print("PDF to DOCX conversion complete.")
```

```python
doc = Document(docx_path)
translator = GoogleTranslator(source='it', target='en')

# Translate each paragraph individually
for para in doc.paragraphs:
    if para.text.strip():
        para.text = translator.translate(para.text)

translated_docx_path = r"C:\Users\Kidia.Gelaye\Downloads\Translated_Formatted.docx"
doc.save(translated_docx_path)

print("Translation of DOCX complete.")
```

```python
translated_pdf_path = r"C:\Users\Kidia.Gelaye\Downloads\Translated_Final.pdf"
convert(translated_docx_path, translated_pdf_path)

print(f"Translated formatted PDF saved at: {translated_pdf_path}")
```

# to capture foot notes and origianl fonts, links..etc !!!! takes more time

```python

pdf_path = r"C:\Users\Kidia.Gelaye\Downloads\Demo.pdf"
docx_path = r"C:\Users\Kidia.Gelaye\Downloads\Original_Formatted.docx"

cv = Converter(pdf_path)
cv.convert(docx_path, start=0, end=None)
cv.close()

print("PDF to DOCX conversion complete.")
```

```python
doc = Document(docx_path)
translator = GoogleTranslator(source='it', target='en')

# Safely translate preserving original formatting (font size, style)
for para in doc.paragraphs:
    if para.runs:
        for run in para.runs:
            original_text = run.text
            if original_text and original_text.strip():
                try:
                    translated_text = translator.translate(original_text)
                    run.text = translated_text
                except Exception as e:
                    print(f"Skipping translation for a problematic run: '{original_text[:20]}...': {e}")

translated_docx_path = r"C:\Users\Kidia.Gelaye\Downloads\Translated_Formatted_Styled.docx"
doc.save(translated_docx_path)

print("Robust translation completed (font size, footnotes, style preserved).")
```

```python
translated_pdf_path = r"C:\Users\Kidia.Gelaye\Downloads\Translated_Final_Styled.pdf"
convert(translated_docx_path, translated_pdf_path)

print(f"Translated, formatted, styled PDF saved at: {translated_pdf_path}")
```
